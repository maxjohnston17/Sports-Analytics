# Hockey Data Exploration

## Potential Chart Idea

```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate two datasets with random numbers
dataset1 = np.random.uniform(low=70, high=95, size=25)
dataset2 = np.random.uniform(low=70, high=95, size=15)

# Step 2: Calculate the mean of each dataset
mean1 = np.mean(dataset1)
mean2 = np.mean(dataset2)

# Step 3: Plot the means on a bar graph
labels = ['Starter', 'Sub']
means = [mean1, mean2]

plt.bar(labels, means, color=['blue', 'orange'])
plt.ylabel('Save Percentage Above Expected')
plt.title('Goaltender Performance in Games with Substitution')
plt.show()
```
